```python
data_w3= data_w[['STORECODE','MONTH','DAY','VALUE']].groupby(['STORECODE','MONTH','DAY']).sum()
data_w3
```

```python
data_w4=data_w3.groupby(['STORECODE','MONTH']).size().reset_index(name='Count')
data_w4
```

```python
data_w5=pd.read_csv('compete/data_working1.csv')
data_w6=pd.merge(data_w5,data_w4, on=['STORECODE','MONTH'])
data_w6
```

```python
data_w6["TOTAL_VALUE_31"]=data_w6['VALUE']/data_w6['Count']*31
data_w6
```

```python
data_w6.drop(['VALUE','Count','QTY'], axis = 1, inplace=True)
data_w6.rename(columns = {'TOTAL_VALUE_31':'VALUE'}, inplace = True)
data_w6
```

```python
data_w7= data_w6.groupby(['GRP','MONTH','STORECODE','VALUE']).count()
data_w7
```

```python
data_w7.to_csv('compete/data_w7.csv') 
```

```python
data_val1= data_val[['GRP','MONTH','STORECODE']].groupby(['GRP','MONTH','STORECODE']).count()
data_val1
```

```python
data_w8=pd.read_csv("compete/data_w7.csv")
data_w9= pd.merge(data_val1,data_w8, on=['GRP','MONTH','STORECODE'], how='outer')
data_w9
```

```python
month = pd.get_dummies(data_w9["MONTH"])
store = pd.get_dummies(data_w9["STORECODE"])
train = pd.concat([data_w9,month,store], axis=1)
train.drop(["GRP","MONTH","STORECODE"],axis=1,inplace=True)
train
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=31)
imputer1=imputer.fit_transform(train)
train['predict']=imputer1[:,0]
```

```python
data_w9["predict"]=train["predict"]
data_w9.drop(['VALUE'], axis = 1, inplace=True)
data_w9.rename(columns = {'predict':'VALUE'}, inplace = True)

data_w9
```

```python
data_w10=data_w9[['STORECODE','MONTH','GRP','VALUE']].groupby(['STORECODE','MONTH','GRP','VALUE']).count()
data_w10
```

```python
data_w10.to_csv("compete/final_filled_KNN.csv")
```

```python
data_w8_2=pd.read_csv("compete/data_w7.csv")
data_w9_2= pd.merge(data_val1,data_w8_2, on=['GRP','MONTH','STORECODE'], how='outer')
data_w9_2
```

```python
month_2 = pd.get_dummies(data_w9_2["MONTH"])
store_2 = pd.get_dummies(data_w9_2["STORECODE"])
grp_2 = pd.get_dummies(data_w9_2["GRP"])
train_2 = pd.concat([data_w9_2,month_2,store_2,grp_2], axis=1)
train_2.drop(["GRP","MONTH","STORECODE"],axis=1,inplace=True)
train_2
```

```python
from sklearn.ensemble import RandomForestRegressor

train_2_1 = train_2[pd.isnull(train_2['VALUE']) == False]
train_2_2 = train_2[pd.isnull(train_2['VALUE'])]

train_2_1.shape
```

```python
train_2_2.shape
```

```python
X_1=train_2_1.iloc[:,1:].values
y_1=train_2_1.iloc[:,0].values

X_1.shape
```

```python
X_2=train_2_2.iloc[:,1:].values
y_2=train_2_2.iloc[:,0].values

X_2.shape
```

```python
rfModel = RandomForestRegressor()
rfModel.fit(X_1, y_1)

generatedValues = rfModel.predict(X_2)
```

```python
train_2_2['VALUE'] = generatedValues
data = train_2_1.append(train_2_2)
data
```

```python
data.sort_index(inplace=True)
data
```

```python
data_w9_2["predict"]=data["VALUE"]
data_w9_2
```

```python
data_w9_2.to_csv('compete2/data_w9_2.csv') 
```
